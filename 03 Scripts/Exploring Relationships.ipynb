##### There also appears to be a positive correlation (about 0.55) between freedom and happiness score.
##### The correlation between trust in government and happiness score, as well as generosity and happiness score both show a weak positive correlation. This could be interpreted as, "The less generosity shown, the less happy" or "The more mistrust shown towards the government, the less happy."

### 3.1b Scatterplots

```python
# scatterplot for social support and happiness score

sns.lmplot(x = 'happiness_score', y = 'social_support', data = df_combined)
```

##### Note: The scatterplot shows that there is definitely a positive correlation between the amount of social support someone receives and their happiness. As social support increases, so does happiness.

```python
# scatterplot for health and happiness score

sns.lmplot(x = 'happiness_score', y = 'health', data = df_combined)
```

##### Note: The scatterplot shows a definite positive correlation between heath and happiness score. Most of the points are along the trend line as well, indicating this strong correlation. However, there are a few lower points.

```python
# scatterplot for freedom and happiness score

sns.lmplot(x = 'happiness_score', y = 'freedom', data = df_combined)
```

##### Note: The scatterplot shows a positive correlation between freedom and happiness score. However, the points are not as tight along the trend line as the correlation between health and happiness, indicating that the correlation is not as strong.

```python
# scatterplot for gdp per capita and happiness score

a=sns.lmplot(x = 'happiness_score', y = 'gdp_per_capita', data = df_combined)
```

##### Note: The scatterplot shows a strong positive correlation between gdp per capita and happiness score. This correlation is quite strong, since most of the data points are close or on top of the trend line.

### 3.1d Categorical Plot

```python
sns.histplot(df_combined['happiness_score'], bins = 20, kde = True)
```

##### Notes: Based off of the histogram, we can create the following categories:
    "Least Happy" <4.25
    "Average Happy" >=4.25 and <6.5
    "Most Happy" >=6.5

```python
sns.histplot(df_combined['health'], bins = 20, kde = True)
```

##### Notes: Based off of the histogram, we can create the following categories:
     "Poor Health" < 0.55
     "Moderate Health" >= 0.55 and < 0.9
     "Excellent Health" >=0.9

```python
sns.histplot(df_combined['gdp_per_capita'], bins = 20, kde = True)
```

##### Notes: Based off of the histogram, we can create the following categories:
     "Low" < 0.75
     "Average" >= 0.75 and < 1.3
     "High" >=1.3

```python
df_combined.loc[df_combined['gdp_per_capita'] < 0.75, 'GDP per capita levels'] = 'Low'
```

```python
df_combined.loc[(df_combined['gdp_per_capita'] >= 0.75) & (df_combined['gdp_per_capita'] < 1.3), 'GDP per capita levels'] = 'Average'
```

```python
df_combined.loc[df_combined['gdp_per_capita'] >= 1.3, 'GDP per capita levels'] = 'High'
```

```python
df_combined['GDP per capita levels'].value_counts(dropna = False)
```

```python
# categorical plot in seaborn using gdp per capita levels above

sns.set(style="ticks")
g = sns.catplot(x="gdp_per_capita", y="happiness_score", hue="GDP per capita levels", data=df_combined)
```
